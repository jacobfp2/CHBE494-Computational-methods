## Question 1

```python
import numpy as np
from numpy.random import rand, randint
import matplotlib.pyplot as plt
def func(n):
    duplicate = 0
    for i in range(0, 5000):
        generator = randint(1, 366, n)
        if len(set(generator)) < len(generator):
            duplicate += 1
        #print(generator)
        #print(duplicate)
    probability = duplicate/5000
    return probability


x = []
y = []
for n in range(40):
    x.append(n)
    y.append(func(n))
    
plt.plot(x,y, color = 'r', lw = 3)
plt.xlabel('N')
plt.ylabel('Probability')
plt.show()
```

## Question 2

```python
def newton(myfun, x, epsilon = 1e-6, delta = 1e-8):
    
    while abs(myfun(x)) > epsilon:
        x = x - myfun(x)/((myfun(x + delta) - myfun(x))/delta)
    return x

def myfun(x):
    return  np.log(x) + x**2

print(newton(myfun,1,1e-6, 1e-8))
```

## Question 3

```python
def secant(myfun, x, epsilon = 1e-6):
    # Create storage for previous x
    # Doesn't matter what it's initialized to since it'll be overrode anyways
    x0 = 0.00001
    while abs(myfun(x)) > epsilon:
        x2 = x - myfun(x)* ((x-x0)/(myfun(x)-myfun(x0)))
        #current x becomes old x
        x0 = x
        #new x is now current x
        x = x2
    return x
print(secant(myfun,1,1e-6))
```

```python
import math as m
def func(t,x,n):
    final = 0
    for i in range(1,n+1):
        final += np.sin((2*i-1)*x)/((2*i-1)**2)*np.exp(-(2*i-1)**2*0.5**2*t)
    final = (8/np.pi)* final
    return final                                                 

#first graph at N = 5
x = np.linspace(0,np.pi,100)
func0 = func(0, x, 5)
func1 = func(1, x, 5)
func2 = func(2, x, 5)
func3 = func(3, x, 5)
func4 = func(4, x, 5)
func5 = func(5, x, 5)
plt.plot(x,func0, color = 'blue', lw = 3, label = 't = 0')
plt.plot(x,func1, color = 'orange', lw = 3, label = 't = 1')
plt.plot(x,func2, color = 'green', lw = 3, label = 't = 2')
plt.plot(x,func3, color = 'red', lw = 3, label = 't = 3')
plt.plot(x,func4, color = 'purple', lw = 3, label = 't = 4')
plt.plot(x,func5, color = 'brown', lw = 3, label = 't = 5')
plt.ylabel('T')
plt.xlabel('X')
plt.legend(fontsize = 12, loc = 'right')
plt.title('N=5')
plt.show()

#second graph at N = 100
fig, ax = plt.subplots()
FUNC0 = func(0, x, 100)
FUNC1 = func(1, x, 100)
FUNC2 = func(2, x, 100)
FUNC3 = func(3, x, 100)
FUNC4 = func(4, x, 100)
FUNC5 = func(5, x, 100)
ax.plot(x,FUNC0, color = 'blue', lw = 3, label = 't = 0')
ax.plot(x,FUNC1, color = 'orange', lw = 3, label = 't = 1')
ax.plot(x,FUNC2, color = 'green', lw = 3, label = 't = 2')
ax.plot(x,FUNC3, color = 'red', lw = 3, label = 't = 3')
ax.plot(x,FUNC4, color = 'purple', lw = 3, label = 't = 4')
ax.plot(x,FUNC5, color = 'brown', lw = 3, label = 't = 5')
plt.ylabel('T')
plt.xlabel('X')
plt.legend(fontsize = 12, loc = 'right')
plt.title('N=100')
plt.show()
```
